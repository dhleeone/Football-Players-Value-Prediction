```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# requsts.get()으로 url정보 요청하기 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
(KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}

player_list=[]

for i in range(1, 9):

    url = f"https://www.futbin.com/players?page={i}&version=tots&sort=Player_Shooting&order=desc"


    r= requests.get(url, headers=headers)
    r.status_code  

    soup = BeautifulSoup(r.text, 'html.parser')  # r.content 대신 r.text도 가능

    player_info= soup.find_all('tr', class_=['player_tr_1', 'player_tr_2'])

    # player_info에서 'td'태그만 모두 찾기
    for info in player_info:
        player = info.find_all("td")

        #number = player[0].text 
        #name = player[3].text 
        NAME= player[0].img['alt'][:-3]
        RAT = player[1].text
        POS = player[2].text
        PAC = player[8].text
        SHO = player[9].text
        PAS = player[10].text
        DRI = player[11].text
        DEF = player[12].text
        PHY = player[13].text
        HEI = player[14].text[2:5]
        IGS = player[17].text
        

        
#         value = value[1:-1]
#         print(value)
        
        player_list.append([NAME, RAT, POS, PAC, SHO, PAS, DRI, DEF, PHY, HEI, IGS])
    
    
    
    time.sleep(1)

    # pd.DataFrame()으로 저장하기

fifa_df = pd.DataFrame(player_list, 
                   columns=['NAME', 'RAT', 'POS', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'HEI', 'IGS'])
fifa_df
```

```python
fifa_df.to_excel('fifa_data.xlsx')
```

```python
fifa_df.to_csv('fifa_data.csv')
```
